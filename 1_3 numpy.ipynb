# Случайные распределения 

```python
# нормальное распределение с центром в нуле и единичной дисперсией

normal = random.normal(size = 100)
normal
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 20)
```

```python
normal = random.normal(size = 10000)
```

```python
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
# сдвинуть центр распределения в точку mu

mu = 10
normal = random.normal(loc = mu, size = 10000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
# изменить стандартное отклонение sigma scale

sigma = 10
normal = random.normal(scale = sigma, size = 10000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
sigma = 1000
normal = random.normal(scale = sigma, size = 10000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
# комбинация аргументов 

mu = 15_000
sigma = 1000
normal = random.normal(loc = mu, scale = sigma, size = 10000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
# задать распределение руками

mu = 10
sigma = 3
normal = random.normal(size = 10000) * sigma + mu
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal, bins = 200)
```

```python
mu = np.array([10, 100, 1000]).reshape(-1, 1)
sigma = np.array([3, 30, 300]).reshape(-1, 1)


normal = random.normal(loc = mu, scale = sigma, size = (3, 10000))
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal[0, :], bins = 200)  # график по каждой строке 
```

```python
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal[1, :], bins = 200)
```

```python
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(normal[2, :], bins = 200)
```

```python
# равномерное распределение на отрезке от 0 до 1

uniform = random.uniform(size = 10000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(uniform, bins = 200)
```

```python
uniform = random.uniform(size = 1000000)
plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(uniform, bins = 200)
```

```python
# равномерное распределение от a до b 

a = -2
b = 2

uniform = random.uniform(low = a, high = b, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(uniform, bins = 200)
```

```python
# треугольное вероятностное распределение

triangular = random.triangular(0, 0.5, 1, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(triangular, bins = 200)
```

```python
triangular = random.triangular(-3, 10, 15, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(triangular, bins = 200)
```

```python
# экспоненциальное распределение 

lambd = 1
exponential = random.exponential(lambd, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(exponential, bins = 200)
```

```python
lambd = 100
exponential = random.exponential(lambd, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(exponential, bins = 200)
```

```python
# распределение Парето (закон Ципфа из лингвистики)

k = 100
pareto = random.pareto(k, size = 1000000)

plt.figure(figsize = (12, 12))
plt.grid()
hist = plt.hist(pareto, bins = 200)
```
