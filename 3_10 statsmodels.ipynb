# Statsmodels. Примеры статистических моделей на python

1. Statsmodel. Ordinary Least Squares

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
plt.style.use('ggplot')
```

```python
nsample = 100
x = np.linspace(0, 10, 100)
X = np.column_stack((x, x ** 3))
beta = np.array([1, 0.1, 10])
e = np.random.normal(size = nsample)

X = sm.add_constant(X)
y = np.dot(X, beta) + e

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
pred_ols = results.get_prediction()
iv_l = pred_ols.summary_frame()['obs_ci_lower']
iv_u = pred_ols.summary_frame()['obs_ci_upper']

fig, ax = plt.subplots(figsize = (8, 6))

ax.plot(x, y, 'o', label = 'data', c = 'm')
#ax.plot(x, y, 'b', label = True)
ax.plot(x, results.fittedvalues, 'r--.', label = 'OLS')
ax.plot(x, iv_u, 'r--')
ax.plot(x, iv_l, 'r--')
ax.legend(loc = 'best')
```

Нелинейный случай 

```python
nsample = 50
sig = 0.5
x = np.linspace(0, 20, nsample)
X = np.column_stack((x, np.sin(x), (x - 5) ** 2, np.ones(nsample)))
beta = [0.5, 0.5, -0.02, 5.0]

y_true = np.dot(X, beta)
y = y_true + sig * np.random.normal(size = nsample)
```

```python
res = sm.OLS(y, X).fit()
print(res.summary())
```

```python
pred_ols = res.get_prediction()
iv_l = pred_ols.summary_frame()['obs_ci_lower']
iv_u = pred_ols.summary_frame()['obs_ci_upper']

fig, ax = plt.subplots(figsize = (8, 6))

ax.plot(x, y, 'o', label = 'data')
ax.plot(x, y_true, 'b-', label = 'True')
ax.plot(x, res.fittedvalues, 'r--.', label = 'OLS')
ax.plot(x, iv_u, 'r--')
ax.plot(x, iv_l, 'r--')
ax.legend(loc = 'best')
```

2. Фильтр Ходрика-Прескотта

```python
dta = sm.datasets.macrodata.load_pandas().data
index = pd.Index(sm.tsa.datetools.dates_from_range('1959Q1', '2009Q3'))
dta.index = index
gdp_cycle, gdp_trend = sm.tsa.filters.hpfilter(dta.realgdp)
```

```python
dta
```

```python
fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111)
dta.realgdp.plot(ax = ax)
legend = ax.legend(loc = 'upper left')
legend.prop.set_size(20)
```

```python
gdp_decomp = dta[['realgdp']].copy()
gdp_decomp['cycle'] = gdp_cycle
gdp_decomp['trend'] = gdp_trend
```

```python
fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111)
gdp_decomp[['trend', 'realgdp']]['2000-03-31':].plot(ax = ax, fontsize = 16)
legend = ax.get_legend()
legend.prop.set_size(20)
```

```python
fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111)
gdp_decomp[['cycle', 'realgdp']]['2000-03-31':].plot(ax = ax, fontsize = 16)
legend = ax.get_legend()
legend.prop.set_size(20)
```
