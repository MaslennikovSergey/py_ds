```python
text = text.replace('\n', ' ').replace('  ', ' ').replace('"','').replace('.', '').replace('   ', ' ').replace(';', '').lower()
text_splited = text.split( )
while '' in text_splited:
    text_splited.remove()
len(text_splited)
```

```python
vocabulary = set(text_splited)
len(vocabulary)
```

```python
frequency_dictionary = {word:text_splited.count(word) for word in vocabulary}
frequency_dictionary
```

```python
temp_dict = frequency_dictionary.copy()

ordered_list_of_words = []

while temp_dict:
    max_num = max(temp_dict.values())
    keys = temp_dict.keys()
    for word in keys:
        if temp_dict[word] == max_num:
            current_word = word
            break
    temp_dict.pop(current_word)
    ordered_list_of_words.append(current_word)

ordered_list_of_words
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

text_splited_lemmatized = [lemmatizer.lemmatize(word) for word in text_splited]
vocabulary_lemmatized = set(text_splited_lemmatized)
len(vocabulary_lemmatized)
```

```python
frequency_dictionary_lemmatized = {word:text_splited_lemmatized.count(word) for word in vocabulary_lemmatized}
frequency_dictionary_lemmatized
```

```python
temp_dict_lemmatized = frequency_dictionary_lemmatized.copy()

ordered_list_of_words_lemmatized = []

while temp_dict:
    max_num = max(temp_dict_lemmatized.values())
    keys = temp_dict_lemmatized.keys()
    for word in keys:
        if temp_dict_lemmatized[word] == max_num:
            current_word = word
            break
    temp_dict_lemmatized.pop(current_word)
    ordered_list_of_words_lemmatized.append(current_word)

ordered_list_of_words_lemmatized
```

```python
from matplotlib import pyplot as plot
x = ordered_list_of_words_lemmatized.copy()
y = [frequency_dictionary_lemmatized[word] for word in x]

x1 = [i for i in range(len(y))]

plt.figure(figsize = (30, 10))
plt.grid(True)
plt.bar(x1, y)
```
