## Библиотека Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset('penguins')
```

```python
sns.set_theme()
```

```python
penguins
```

```python
plt.figure(figsize = (13, 13))
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = 'species', multiple = 'stack')
```

```python
plt.figure(figsize = (13, 13))
sns.kdeplot(data = penguins, x = 'flipper_length_mm', hue = 'species', multiple = 'stack')
```

```python
sns.displot(data = penguins, x = 'flipper_length_mm', hue = 'species', col= 'species')
```

```python
sns.displot(data = penguins, x = 'flipper_length_mm', hue = 'species', col= 'species', kind='kde')
```

```python
plt.figure(figsize = (13, 13))
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'species')
```

```python
plt.figure(figsize = (13, 13))
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'island')
```

```python
plt.figure(figsize = (13, 13))
sns.scatterplot(data = penguins, x = 'species', y = 'island')
```

```python
plt.figure(figsize = (13, 13))
sns.histplot(data = penguins, x = 'species', hue = 'species', shrink = .8, alpha = .8, legend = False)
```

```python
sns.jointplot(data = penguins, x = 'flipper_length_mm', y = 'bill_length_mm', hue = 'species')
```

```python
sns.pairplot(data = penguins, hue = 'species')
```
